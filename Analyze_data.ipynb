# read multi file csv

```
import os
import pandas as pd
def getListProcess(folder_name):
  df = pd.DataFrame() 
  for root, dir, file in os.walk(folder_name):
    for f in file:
      if '.csv' in f:
        print(os.path.join(root, f))
        df = pd.concat([df, pd.read_csv(os.path.join(root, f))], sort=False)
    return df   
```

```
data = pd.DataFrame() 
for root, dir, file in os.walk("/content/drive/My Drive/Data_2021/Data_9-2-2021"):
  for d in dir:
    df_process=  getListProcess(os.path.join(root, d) )
    data = pd.concat([data, df_process], sort=False)
# data.sample(5)
```
